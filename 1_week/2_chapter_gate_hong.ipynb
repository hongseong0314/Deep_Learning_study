```python
import numpy as np
```

**AND**

```python
def AND(x, y):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x * w1 + y * w2
    return 0 if tmp <= theta else 1
```

```python
print(AND(0, 0))
print(AND(1, 0))
print(AND(0, 1))
print(AND(1, 1))
```

```python
# 가중치 편향 포함한 AND
def AND(x, y):
    x_ = np.array([x, y])
    w = np.array([0.5, 0.5])
    b = -0.7
    tem = np.sum(x_ * w) + b
    return 0 if tem <= 0 else 1
```

**NAND**

```python
def NAND(x1, x2):
    x_ = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tem = np.sum(x_ * w) + b
    return 0 if tem <= 0 else 1
```

**OR**


```python
def OR(x1, x2):
    x_ = np.array([x1, x2])
    w = np.array([1., 1.])
    b = 0.0
    tem = np.sum(x_ *  w) + b
    return 0 if tem <= 0 else 1    
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

**XOR**

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```
